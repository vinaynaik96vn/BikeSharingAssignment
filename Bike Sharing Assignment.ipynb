# Step 1 :- Reading and Understanding the data.

Dropping unnecessary columns.

```python
#since instant is an index column we can drop it
bike.drop(["instant"], axis = 1, inplace = True)
bike.head()
```

Mapping Variables "season", "month", "weathersit", "weekday"

```python
bike["season"] = bike.season.map({1:"spring", 2:"summer", 3:"fall", 4:"winter"})
bike["month"] = bike.month.map({1:"Jan", 2:"Feb", 3:"March", 4:"April", 5:"May", 6:"June", 7:"July", 8:"Aug", 9:"Sept", 10:"Oct", 11:"Nov", 12:"Dec"})
bike["weathersit"] = bike.weathersit.map({1:"Clear", 2:"Mist + Cloudy", 3:"Light Snow", 4:"Heavy Rain"})
bike["weekday"] = bike.weekday.map({0:"Sunday", 1:"Monday", 2:"Tuesday", 3:"Wednesday", 4:"Thusday", 5:"Friday", 6:"Saturday"})
bike.head()
```

### Dividing into X and Y sets for model building.

# Step 6 :- Building a Linear Model

```python
#Importing RFE and Linear Regression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#Running RFE with output number of variables equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building Model using statsmodel for detailed summary.

```python
#creating X_test dataframe with RFE selected variables.
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm
```

```python
#Adding a constant variable
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit()
```

```python
print(lm.summary())
```

### Checking VIF

```python
#Drop the constant term
X_train_rfe = X_train_rfe.drop("const", axis = 1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#Calculate the VIF for new model
vif = pd.DataFrame()
X = X_train_rfe
vif["Features"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
vif["VIF"] = round (vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Saturday is insignificant in presence of other variable due to high pvalue and low VIF it can be dropped.

```python
#dropping Saturday
X_train_new1 = X_train_rfe.drop(["Saturday"], axis = 1)
```

```python
#Rebuilding model without "Saturday"
X_train_lm = sm.add_constant(X_train_new1)

lm = sm.OLS(y_train, X_train_lm).fit()

print(lm.summary())
```

```python
X_train_lm1 = X_train_lm.drop(["const"], axis = 1)
```

```python
vif = pd.DataFrame()
X = X_train_new1
vif["Features"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
vif["VIF"] = round (vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Dropping variable "humidity" as it is having high VIF.


```python
#dropping humidity
X_train_new2 = X_train_lm1.drop(["humidity"], axis = 1)
```

```python
#Rebuilding model without "humidity"
X_train_lm = sm.add_constant(X_train_new2)

lm2 = sm.OLS(y_train, X_train_lm).fit()

print(lm.summary())
```

```python
X_train_lm2 = X_train_lm.drop(["const"], axis = 1)
```

```python
vif = pd.DataFrame()
X = X_train_new2
vif["Features"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
vif["VIF"] = round (vif["VIF"], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### The p-values and VIFs both are in acceptable range. So we can say that our model is ready for prediction.

-  The p-value for each term tests the null hypothesis that the coefficient is equal to zero (no effect). A low p-value (<0.05) indicates that you can reject the null hypothesis.
-  A rule of thumb commonly used in practice is if a VIF is > 10, you have high multicollinearity. In our case, with values less than 5, we are in good shape, and can proceed with our regression.
-  R-squared measures the strength of the relationship between your model and the dependent variable on a convenient 0 – 100% scale. And we have the R-square value of 0.836 or 83.6%.
-  The adjusted R-squared adjusts for the number of terms in the model. And we got it around 0.832 or 83.2%.
So our model is lm2 which is obtained by removing "Saturday" and "humidity" variables from the RFE support columns.
